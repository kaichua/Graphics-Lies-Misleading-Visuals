```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('total_unemployed.csv')
df['Year'] = pd.to_datetime(df['Year'], format='%b-%y')
df.sort_values(by='Year', ascending = True, inplace = True)
df.rename({'Number of unemployed (thousands)' : 'Total'}, axis=1 , inplace = True)
df['Year'] = df['Year'].apply(lambda x: x.strftime('%b-%Y'))
df.set_index('Year', inplace=True)
```

```python
plt.figure(figsize=(15,10))
plt.plot(df.index, df.Total, linewidth=5)
plt.plot(df.index, df.Total, 'o', markersize=20, color='black')

plt.xticks(fontsize=15, fontweight='bold', rotation=70)
plt.yticks(fontsize=15, fontweight='bold')
plt.ylabel('(thousands)',fontsize=20, fontweight='bold')
plt.title('Total Unemployed',fontsize=20, fontweight='bold')

plt.tight_layout()

plt.savefig('Figure_4.png')
```

```python
df_selected = df[df.index.isin(['Dec-2007','Sep-2008','Mar-2009','Jun-2010'])]

plt.figure(figsize=(15,10))
plt.plot(df_selected.index, df_selected.Total, linewidth=5)
plt.plot(df_selected.index, df_selected.Total, 'o', markersize=20, color='black')

plt.xticks(fontsize=15, fontweight='bold')
plt.yticks(fontsize=15, fontweight='bold')
plt.ylabel('(thousands)',fontsize=20, fontweight='bold')
plt.title('Selected Total Unemployed',fontsize=20, fontweight='bold')

for i in range(len(df_selected.index)):    
    plt.text(i+0.1, df_selected.iloc[i]['Total'], str((df_selected.iloc[i]['Total'])), fontsize=20)

plt.tight_layout()

plt.savefig('Figure_5.png')
```
